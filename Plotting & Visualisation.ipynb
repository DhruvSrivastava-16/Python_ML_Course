```python
x=np.array([0,1,2])
y=np.arange(1,10,3)
y2=np.arange(11,21,4)
```

```python
print('y is: ',y)
print('y2 is: ',y2)
```

```python
plt.bar(x+0.5,y,width=0.25,label="one",tick_label=[1,2,3])
plt.bar(x+1.75,y2,width=0.15,label="two",tick_label=['a','b','c'])
plt.title('First Bar Graph!')
#plt.ylim(0,10)
plt.legend()
```
